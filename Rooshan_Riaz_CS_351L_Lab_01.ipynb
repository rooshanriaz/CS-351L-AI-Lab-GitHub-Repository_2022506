```python
CS 351L - AI Lab
Instructor: Mr. Usama Arshad, PhD CS
Student Name: Rooshan Riaz
Registration No: 2022506
BS Cybersecurity - 5th Semester

Non-AI Version: Player vs Computer

import random

def number_guessing_game():
    # The computer selects a random number between 1 and 100
    number_to_guess = random.randint(1, 100)
    attempts = 0
    max_attempts = 10  # Player has 10 attempts to guess the number

    print("Welcome to the Number Guessing Game!")
    print("Guess a number between 1 and 100. You have 10 attempts.")

    # Loop for the player to make guesses
    while attempts < max_attempts:
        guess = int(input("Enter your guess: "))
        attempts += 1

        if guess < number_to_guess:
            print("Too low!")
        elif guess > number_to_guess:
            print("Too high!")
        else:
            print(f"Congratulations! You guessed the number in {attempts} attempts.")
            return

    print(f"Sorry, you've used all your attempts. The number was {number_to_guess}.")

# Run the game
number_guessing_game()

AI Version: Computer vs Player

def ai_number_guessing_game():
    # Player selects a number
    print("Think of a number between 1 and 100, and I (the AI) will try to guess it.")
    low = 1
    high = 100
    attempts = 0

    # Loop until the AI guesses the number correctly
    while low <= high:
        guess = (low + high) // 2  # AI uses binary search to guess
        attempts += 1

        print(f"AI's guess is: {guess}")
        feedback = input("Enter 'h' if too high, 'l' if too low, or 'c' if correct: ").lower()

        if feedback == 'c':
            print(f"I (AI) guessed the number in {attempts} attempts!")
            return
        elif feedback == 'h':
            high = guess - 1  # If too high, reduce the upper bound
        elif feedback == 'l':
            low = guess + 1  # If too low, increase the lower bound

    print("Something went wrong!")

# Run the AI version
ai_number_guessing_game()

Explanation of Both Versions:

Non-AI Version: The player guesses the number while the computer provides feedback. Itâ€™s a simple interaction where the player uses logic to adjust their guesses based on whether the guess is too high or too low.
AI Version: The computer plays the guessing game using a binary search strategy to guess the player's number. Binary search is an efficient algorithm that repeatedly divides the search range in half.

Breadth-First Search (BFS) Version

from collections import deque

def bfs_number_guessing_game():
    print("Think of a number between 1 and 100, and I (the AI) will try to guess it using BFS.")

    low = 1
    high = 100
    mid = (low + high) // 2

    queue = deque([mid])
    attempts = 0
    visited = set()

    while queue:
        guess = queue.popleft()
        attempts += 1
        if guess in visited:
            continue  # Skip already guessed numbers
        visited.add(guess)

        print(f"AI's guess is: {guess}")
        feedback = input("Enter 'h' if too high, 'l' if too low, or 'c' if correct: ").lower()

        if feedback == 'c':
            print(f"I (AI) guessed the number in {attempts} attempts using BFS!")
            return
        elif feedback == 'h':
            for i in range(guess-1, low-1, -1):
                if i not in visited:
                    queue.append(i)
        elif feedback == 'l':
            for i in range(guess+1, high+1):
                if i not in visited:
                    queue.append(i)

    print("Something went wrong!")

# Run the BFS version
bfs_number_guessing_game()

Explanation of BFS Version:

BFS explores numbers in a breadth-first manner. It starts by guessing the middle value and then checks smaller or larger values in a breadth-first manner by enqueuing neighbors of the current guess.
We use a queue (deque) to store the guesses and explore numbers level by level. The algorithm gradually refines its guesses based on feedback. This code implements a number guessing game using
Breadth-First Search (BFS), where the AI tries to guess a number between 1 and 100. The AI starts by guessing the middle number and then adjusts its guesses based on player feedback ('h' for too high,
'l' for too low, 'c' for correct). The guesses are stored in a queue, and the AI systematically explores all unvisited numbers in the specified range. If the guess is incorrect, numbers higher or lower
than the current guess are added to the queue for further exploration. The AI continues until it finds the correct number or exhausts the que

Depth-First Search (DFS) Version:

def dfs_number_guessing_game():
    print("Think of a number between 1 and 100, and I (the AI) will try to guess it using DFS.")

    low = 1
    high = 100
    attempts = 0
    stack = [(low, high)]  # Stack maintains the range to explore

    while stack:
        low, high = stack.pop()
        if low > high:
            continue

        guess = (low + high) // 2
        attempts += 1

        print(f"AI's guess is: {guess}")
        feedback = input("Enter 'h' if too high, 'l' if too low, or 'c' if correct: ").lower()

        if feedback == 'c':
            print(f"I (AI) guessed the number in {attempts} attempts using DFS!")
            return
        elif feedback == 'h':
            stack.append((low, guess - 1))  # Explore the lower range
        elif feedback == 'l':
            stack.append((guess + 1, high))  # Explore the higher range

    print("Something went wrong!")

# Run the DFS version
dfs_number_guessing_game()

Explanation fo DFS Version:

DFS uses a stack to explore deeply before backtracking. It starts with the middle number and proceeds in one direction, either towards lower or higher numbers, until the correct number is found.
If the AI receives feedback that the guess is too high, it pushes the lower range onto the stack for further exploration. If it's too low, it explores the higher range.  The game starts by
initializing a stack with the full range of possible numbers (1 to 100). The AI will make guesses from this range, aiming to narrow it down.  The AI uses a stack to keep track of unexplored
ranges. The algorithm pops a range from the stack, makes a guess, and narrows the range until it finds the correct number. When the AI guesses correctly, it prints the number of attempts taken
to find the correct answer. If the stack becomes empty without finding the correct guess, it indicates an error.

Simulated Annealing Version:

import random
import math

def simulated_annealing_number_guessing_game():
    print("Think of a number between 1 and 100, and I (the AI) will try to guess it using Simulated Annealing.")

    low = 1
    high = 100
    attempts = 0
    temperature = 100.0  # Start with a high temperature

    while temperature > 1:
        guess = random.randint(low, high)  # Randomly guess within the range
        attempts += 1

        print(f"AI's guess is: {guess}")
        feedback = input("Enter 'h' if too high, 'l' if too low, or 'c' if correct: ").lower()

        if feedback == 'c':
            print(f"I (AI) guessed the number in {attempts} attempts using Simulated Annealing!")
            return
        elif feedback == 'h':
            high = guess - 1
        elif feedback == 'l':
            low = guess + 1

        temperature *= 0.9  # Gradually cool down the temperature (decrease randomness)

    print("AI couldn't guess the number within the cooling schedule.")

# Run the Simulated Annealing version
simulated_annealing_number_guessing_game()

Explanation of Simulated Annealing Version:

The Simulated Annealing version of the number guessing game allows the AI to guess the player's number by introducing a controlled randomness that decreases
over time. The name comes from the analogy to the annealing process in metallurgy, where a material is heated and then cooled slowly to remove defects and
reach a stable structure. At the start, the AI is in the explorationphase. It makes random guesses anywhere within therange (1 to 100), allowing for maximum
randomness. Based on the feedback, the AI starts adjusting the range it guesses in. If the guess is too high, it lowers the
upper bound, and if it's too low, it raises the lower bound.

Why Simulated Annealing vs. Other Algorithms?

BFS and DFS systematically explore the search space. Simulated Annealing introduces controlled randomness, which can sometimes be more efficient for games
where you want some unpredictability or don't want to always follow a strict pattern. In a number guessing game, SA can guess randomly at first, which introduces
an element of surprise and flexibility. This is more reflective of human guessing, where we might guess wildly at first before narrowing down based on feedback.


```
